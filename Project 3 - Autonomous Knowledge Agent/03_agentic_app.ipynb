### Demonstration 4: Memory Retrieval and Personalization

This demonstrates long-term memory retrieval and how historical context is used.


````python
# Import memory functions
from agentic.memory import get_conversation_history, get_user_preferences, get_historical_context

# Get conversation history for a user
print("=== Conversation History ===")
history = get_conversation_history(user_id="a4ab87", limit=5)
for ticket in history:
    print(f"\nTicket ID: {ticket['ticket_id']}")
    print(f"Status: {ticket['status']}")
    print(f"Issue Type: {ticket.get('issue_type', 'N/A')}")
    print(f"Messages: {len(ticket['messages'])}")
    if ticket['messages']:
        print(f"  First message: {ticket['messages'][0]['content'][:50]}...")

print("\n=== User Preferences ===")
prefs = get_user_preferences(user_id="a4ab87")
print(f"Resolved Issues: {len(prefs['resolved_issues'])}")
print(f"Common Issues: {prefs['common_issues']}")

print("\n=== Historical Context for Login Issue ===")
context = get_historical_context(user_id="a4ab87", current_issue_type="login")
print(context if context else "No historical context available")

# Now test with a new login question to see personalization
print("\n=== Testing Personalization ===")
send_message(
    orchestrator,
    "I'm having login issues again",
    ticket_id="memory-test-1",
    user_id="a4ab87",  # Same user - should use historical context
    account_id="cultpass"
)
````

### Demonstration 5: Logging and Observability

This demonstrates structured logging with routing decisions, classifications, tool usage, and outcomes.


````python
import json
import os
from pathlib import Path
from datetime import datetime

# Find the most recent log file
log_dir = Path("logs")
if log_dir.exists():
    log_files = sorted(log_dir.glob("udahub_*.jsonl"), key=os.path.getmtime, reverse=True)
    
    if log_files:
        latest_log = log_files[0]
        print(f"Reading logs from: {latest_log}\n")
        print("=" * 80)
        
        # Read and display sample log entries
        with open(latest_log, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            
        print(f"Total log entries: {len(lines)}\n")
        print("Sample log entries showing key events:\n")
        print("-" * 80)
        
        # Show different types of log entries
        shown_types = set()
        for line in lines[-50:]:  # Last 50 entries
            try:
                log_entry = json.loads(line)
                log_type = log_entry.get("message", "")
                
                # Show one of each important type
                if "Supervisor routing decision" in log_type and "routing_decision" not in shown_types:
                    print("üìä SUPERVISOR ROUTING DECISION:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("routing_decision")
                
                elif "Ticket classified successfully" in log_type and "classification" not in shown_types:
                    print("üè∑Ô∏è  CLASSIFICATION:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("classification")
                
                elif "Tool invoked" in log_type and "tool_invoke" not in shown_types:
                    print("üîß TOOL INVOCATION:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("tool_invoke")
                
                elif "KB search completed with results" in log_type and "kb_search" not in shown_types:
                    print("üìö KNOWLEDGE BASE SEARCH:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("kb_search")
                
                elif "Resolution attempted" in log_type and "resolution" not in shown_types:
                    print("‚úÖ RESOLUTION ATTEMPT:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("resolution")
                
                elif "Ticket escalated" in log_type and "escalation" not in shown_types:
                    print("üö® ESCALATION:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("escalation")
                
                elif "Conversation saved to database" in log_type and "persistence" not in shown_types:
                    print("üíæ DATABASE PERSISTENCE:")
                    print(json.dumps(log_entry, indent=2))
                    print("-" * 80)
                    shown_types.add("persistence")
                
            except json.JSONDecodeError:
                continue
        
        if not shown_types:
            print("No structured log entries found. Run some conversations first.")
    else:
        print("No log files found. Run some conversations to generate logs.")
else:
    print("Logs directory not found. Run some conversations to generate logs.")
````
